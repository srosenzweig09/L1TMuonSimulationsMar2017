```python
#maxEvents = 400000
maxEvents = 2000000

kDT, kCSC, kRPC, kGEM, kME0 = 0, 1, 2, 3, 4

nlayers = 16  # 5 (CSC) + 4 (RPC) + 3 (GEM) + 4 (DT)

# Decide EMTF hit layer number
class EMTFLayer(object):
  def __init__(self):
    lut = np.zeros((5,5,5), dtype=np.int32) - 99  # (type, station, ring) -> layer
    lut[1,1,4] = 0  # ME1/1a
    lut[1,1,1] = 0  # ME1/1b
    lut[1,1,2] = 1  # ME1/2
    lut[1,1,3] = 1  # ME1/3
    lut[1,2,1] = 2  # ME2/1
    lut[1,2,2] = 2  # ME2/2
    lut[1,3,1] = 3  # ME3/1
    lut[1,3,2] = 3  # ME3/2
    lut[1,4,1] = 4  # ME4/1
    lut[1,4,2] = 4  # ME4/2
    lut[2,1,2] = 5  # RE1/2
    lut[2,1,3] = 5  # RE1/3
    lut[2,2,2] = 6  # RE2/2
    lut[2,2,3] = 6  # RE2/3
    lut[2,3,1] = 7  # RE3/1
    lut[2,3,2] = 7  # RE3/2
    lut[2,3,3] = 7  # RE3/3
    lut[2,4,1] = 8  # RE4/1
    lut[2,4,2] = 8  # RE4/2
    lut[2,4,3] = 8  # RE4/3
    lut[3,1,1] = 9  # GE1/1
    lut[3,2,1] = 10 # GE2/1
    lut[4,1,1] = 11 # ME0
    lut[0,1,1] = 12 # MB1
    lut[0,2,1] = 13 # MB2
    lut[0,3,1] = 14 # MB3
    lut[0,4,1] = 15 # MB4
    self.lut = lut

  def __call__(self, hit):
    index = (hit.type, hit.station, hit.ring)
    entry = self.lut[index]
    return entry

find_emtf_layer = EMTFLayer()

# Decide EMTF legit hit
def is_emtf_legit_hit(hit):
  def check_bx(hit):
    if hit.type == kCSC:
      return hit.bx in (-1,0)
    elif hit.type == kDT:
      return hit.bx in (-1,0)
    else:
      return hit.bx == 0
  def check_emtf_phi(hit):
    if hit.type == kME0:
      return hit.emtf_phi > 0
    elif hit.type == kDT:
      return hit.emtf_phi > 0
    else:
      return True
  #return check_bx(hit) and check_emtf_phi(hit)
  return check_emtf_phi(hit)
```

```python
def find_endsec(endcap, sector):
  endsec = (sector - 1) if endcap == 1 else (sector - 1 + 6)
  return endsec

def calculate_d0(invPt, phi, xv, yv, B=3.811):
  _invPt = np.asarray(invPt, dtype=np.float64)   # needs double precision
  _invPt = np.where(np.abs(_invPt) < 1./10000, np.sign(_invPt+1e-15) * 1./10000, _invPt)
  _R = -1.0 / (0.003 * 3.811 * _invPt)           # R = -pT/(0.003 q B)  [cm]
  _xc = xv - (_R * np.sin(phi))                  # xc = xv - R sin(phi)
  _yc = yv + (_R * np.cos(phi))                  # yc = yv + R cos(phi)
  _d0 = _R - (np.sign(_R) * np.hypot(_xc, _yc))  # d0 = R - sign(R) * sqrt(xc^2 + yc^2)
  return _d0.astype(np.float32, casting='same_kind')
```

```python
class Collector(object):
  def __init__(self, select_fn_1, select_fn_2, coarse_1=1, coarse_2=1):
    self.select_fn_1 = select_fn_1
    self.select_fn_2 = select_fn_2
    self.coarse_1 = coarse_1
    self.coarse_2 = coarse_2

  def run(self, hits):
    selected_hits_1 = np.empty((12,), dtype=np.object)
    selected_hits_2 = np.empty((12,), dtype=np.object)
    for ind in np.ndindex(selected_hits_1.shape):
      selected_hits_1[ind] = []
      selected_hits_2[ind] = []

    # Loop over hits
    for ihit, hit in enumerate(hits):
      hit.endsec = find_endsec(hit.endcap, hit.sector)
      hit.lay = find_emtf_layer(hit)
      assert(hit.lay != -99)

      if self.select_fn_1(hit):
        selected_hits_1[hit.endsec].append(ihit)
      if self.select_fn_2(hit):
        selected_hits_2[hit.endsec].append(ihit)

    # Loop over sectors
    dphis = []
    dthetas = []

    for endsec in xrange(12):
      hit_1 = None
      hit_2 = None

      if selected_hits_1[endsec]:
        ihit = np.random.choice(selected_hits_1[endsec])
        hit_1 = hits[ihit]
      if selected_hits_2[endsec]:
        ihit = np.random.choice(selected_hits_2[endsec])
        hit_2 = hits[ihit]

      if hit_1 is not None and hit_2 is not None:
        # delta phi
        if self.coarse_1 == 1 and self.coarse_2 == 1:
          dphi = hit_1.emtf_phi - hit_2.emtf_phi
        else:
          emtf_phi_1 = np.int32(hit_1.emtf_phi)
          emtf_phi_2 = np.int32(hit_2.emtf_phi)
          emtf_phi_1 //= self.coarse_1
          emtf_phi_2 //= self.coarse_2
          emtf_phi_1 *= self.coarse_1
          emtf_phi_2 *= self.coarse_2
          dphi = emtf_phi_1 - emtf_phi_2
        dphis.append(dphi)
        # delta theta
        dtheta = hit_1.emtf_theta - hit_2.emtf_theta
        dthetas.append(dtheta)
    
    # Return one pair of (dphi, dtheta)
    if dphis:
      dphi = np.random.choice(dphis)
      dtheta = np.random.choice(dthetas)
    else:
      dphi = np.nan
      dtheta = np.nan
    return (dphi, dtheta)
```

```python
# Main analysis

def analysis(verbose=1, omtf=False):
  if omtf:
    tree = load_pgun_omtf()
  else:
    tree = load_pgun()

  select_fn_11 = lambda hit: (hit.type == kCSC and hit.station == 1 and (hit.ring == 1 or hit.ring == 4) and hit.neighbor == 0)
  select_fn_12 = lambda hit: (hit.type == kCSC and hit.station == 1 and (hit.ring == 2 or hit.ring == 3) and hit.neighbor == 0)
  select_fn_2 = lambda hit: (hit.type == kCSC and hit.station == 2 and hit.neighbor == 0)
  select_fn_3 = lambda hit: (hit.type == kCSC and hit.station == 3 and hit.neighbor == 0)
  select_fn_4 = lambda hit: (hit.type == kCSC and hit.station == 4 and hit.neighbor == 0)

  # For CSC stations 1&2, even chambers are 'F', odd chambers are 'R'
  # The GEM-CSC lever arm is longer for the 'R' chambers, shorter for the 'F' chambers
  select_fn_ge11_even = lambda hit: (hit.type == kGEM and hit.station == 1 and hit.ring == 1 and hit.neighbor == 0 and (hit.chamber%2) == 0)
  select_fn_ge11_odd = lambda hit: (hit.type == kGEM and hit.station == 1 and hit.ring == 1 and hit.neighbor == 0 and (hit.chamber%2) == 1)
  select_fn_me11_even = lambda hit: (hit.type == kCSC and hit.station == 1 and (hit.ring == 1 or hit.ring == 4) and hit.neighbor == 0 and (hit.chamber%2) == 0)
  select_fn_me11_odd = lambda hit: (hit.type == kCSC and hit.station == 1 and (hit.ring == 1 or hit.ring == 4) and hit.neighbor == 0 and (hit.chamber%2) == 1)

  collect_0 = Collector(select_fn_11, select_fn_2)
  collect_1 = Collector(select_fn_11, select_fn_3)
  collect_2 = Collector(select_fn_11, select_fn_4)
  collect_3 = Collector(select_fn_12, select_fn_2)
  collect_4 = Collector(select_fn_12, select_fn_3)
  collect_5 = Collector(select_fn_12, select_fn_4)
  collect_6 = Collector(select_fn_2, select_fn_3)
  collect_7 = Collector(select_fn_2, select_fn_4)
  collect_8 = Collector(select_fn_3, select_fn_4)

  collect_9 = Collector(select_fn_ge11_even, select_fn_me11_even)
  collect_10 = Collector(select_fn_ge11_odd, select_fn_me11_odd)
  collect_11 = Collector(select_fn_ge11_even, select_fn_me11_even, coarse_1=1, coarse_2=2)
  collect_12 = Collector(select_fn_ge11_odd, select_fn_me11_odd, coarse_1=1, coarse_2=2)
  collect_13 = Collector(select_fn_ge11_even, select_fn_me11_even, coarse_1=1, coarse_2=4)
  collect_14 = Collector(select_fn_ge11_odd, select_fn_me11_odd, coarse_1=1, coarse_2=4)

  out_part, out_dphis, out_dthetas = [], [], []

  # Loop over events
  for ievt, evt in enumerate(tree):
    if maxEvents != -1 and ievt == maxEvents:
      break

    if verbose and (ievt % 10000 == 0):  print("Processing event: {0}".format(ievt))

    # Skip events without particles
    if len(evt.particles) == 0:
      continue

    # Skip events with very few hits
    if not len(evt.hits) >= 2:
      continue

    # Skip events without ME1 hits
    has_ME1 = False
    for ihit, hit in enumerate(evt.hits):
      if hit.type == kCSC and hit.station == 1:
        has_ME1 = True
        break
      elif hit.type == kME0 and hit.station == 1:
        has_ME1 = True
        break
      elif hit.type == kDT and (hit.station == 1 or hit.station == 2):
        has_ME1 = True
        break
    #if not has_ME1:
    #  continue

    part = evt.particles[0]  # particle gun
    part.invpt = np.true_divide(part.q, part.pt)
    part.d0 = calculate_d0(part.invpt, part.phi, part.vx, part.vy)

    # Skip pathological displaced muons
    if np.hypot(part.vx, part.vy) >= 150:
      continue

    # Skip low-pT muons
    if part.pt < 20:
      continue

    dphi_0, dtheta_0 = collect_0.run(evt.hits)
    dphi_1, dtheta_1 = collect_1.run(evt.hits)
    dphi_2, dtheta_2 = collect_2.run(evt.hits)
    dphi_3, dtheta_3 = collect_3.run(evt.hits)
    dphi_4, dtheta_4 = collect_4.run(evt.hits)
    dphi_5, dtheta_5 = collect_5.run(evt.hits)
    dphi_6, dtheta_6 = collect_6.run(evt.hits)
    dphi_7, dtheta_7 = collect_7.run(evt.hits)
    dphi_8, dtheta_8 = collect_8.run(evt.hits)
    dphi_9, dtheta_9 = collect_9.run(evt.hits)
    dphi_10, dtheta_10 = collect_10.run(evt.hits)
    dphi_11, dtheta_11 = collect_11.run(evt.hits)
    dphi_12, dtheta_12 = collect_12.run(evt.hits)
    dphi_13, dtheta_13 = collect_13.run(evt.hits)
    dphi_14, dtheta_14 = collect_14.run(evt.hits)

    dphis = (dphi_0, dphi_1, dphi_2, dphi_3, dphi_4, dphi_5, dphi_6, dphi_7, dphi_8,
             dphi_9, dphi_10, dphi_11, dphi_12, dphi_13, dphi_14)
    dthetas = (dtheta_0, dtheta_1, dtheta_2, dtheta_3, dtheta_4, dtheta_5, dtheta_6, dtheta_7, dtheta_8,
               dtheta_9, dtheta_10, dtheta_11, dtheta_12, dtheta_13, dtheta_14)

    # Output
    out_part.append((part.invpt, part.eta, part.phi, part.d0))
    out_dphis.append(dphis)
    out_dthetas.append(dthetas)
    continue  # end loop over events

  if verbose: print "DONE"
  return np.asarray(out_part, dtype=np.float32), np.asarray(out_dphis, dtype=np.float32), np.asarray(out_dthetas, dtype=np.float32)
```

```python
from scipy.optimize import curve_fit

def gaus(x,a,mu,sig):
  return a*np.exp(-0.5*np.square((x-mu)/sig))

def fit_gaus(hist, edges, mu=0., sig=1.):
  hist = hist.astype(np.float64)
  edges = edges.astype(np.float64)
  xdata = (edges[1:] + edges[:-1])/2
  ydata = hist
  popt, pcov = curve_fit(gaus, xdata, ydata, p0=[np.max(hist),mu,sig])
  if not np.isfinite(pcov).all():
    raise Exception('Fit has failed to converge.')
  popt[2] = np.abs(popt[2]) # take absolute value of sigma 
  return popt
```

```python
convert_to_rad = lambda x: x * np.deg2rad(1./60)

convert_to_mrad = lambda x: x * np.deg2rad(1000./60)

#pt_slices = [2, 3, 5, 10, 20, 50]
d0_slices = [-150, -80, -50, -20, 20, 50, 80, 150]

#colors = ['C0', 'C1', 'C2', 'C3', 'C4', 'C5']
colors = ['C0', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6']

part_invpt = out_part[:, 0]
part_q = -1 + 2 * np.greater_equal(out_part[:, 0], 0)
part_q = part_q.astype(np.float32)
part_eta = out_part[:, 1]
part_phi = out_part[:, 2]
part_d0 = out_part[:, 3]
```

```python
# Plot deflection angles for |eta| > 1.6, pT > 20 GeV
fig, axs = plt.subplots(3, 2, figsize=(8,12), tight_layout=True)

xlabels = [
  r'$(\phi_{\mathrm{ME1/1}} - \phi_{\mathrm{ME2}}) \cdot q$ [mrad]',
  r'$(\phi_{\mathrm{ME1/1}} - \phi_{\mathrm{ME3}}) \cdot q$ [mrad]',
  r'$(\phi_{\mathrm{ME1/1}} - \phi_{\mathrm{ME4}}) \cdot q$ [mrad]',
  r'$(\phi_{\mathrm{ME3}} - \phi_{\mathrm{ME2}}) \cdot q$ [mrad]',
  r'$(\phi_{\mathrm{ME4}} - \phi_{\mathrm{ME2}}) \cdot q$ [mrad]',
  r'$(\phi_{\mathrm{ME4}} - \phi_{\mathrm{ME3}}) \cdot q$ [mrad]',
]

ylabel = r'entries'

for i in xrange(6):
  for j in xrange(len(d0_slices)-1):
    xlabel = xlabels[i]
    d0_low = d0_slices[j]
    d0_high = d0_slices[j+1]
    color = colors[j]

    sel = (d0_low <= (part_d0 * part_q)) & \
          ((part_d0 * part_q) <= d0_high) & \
          (np.abs(1.0/part_invpt) > 20.) & \
          (np.abs(part_eta) > 1.6)

    if i == 0:
      dphi = convert_to_mrad(out_dphis[:, 0][sel])
      dphi *= part_q[sel]
    elif i == 1:
      dphi = convert_to_mrad(out_dphis[:, 1][sel])
      dphi *= part_q[sel]
    elif i == 2:
      dphi = convert_to_mrad(out_dphis[:, 2][sel])
      dphi *= part_q[sel]
    elif i == 3:
      dphi = -convert_to_mrad(out_dphis[:, 6][sel])
      dphi *= part_q[sel]
    elif i == 4:
      dphi = -convert_to_mrad(out_dphis[:, 7][sel])
      dphi *= part_q[sel]
    elif i == 5:
      dphi = -convert_to_mrad(out_dphis[:, 8][sel])
      dphi *= part_q[sel]

    dphi = dphi[np.isfinite(dphi)]  # remove NaN

    ax = axs[i%3,i/3]
    # For making gaussian fits
    hist, edges, _ = ax.hist(dphi, bins=600, range=(-300,300), histtype='step', color=color, alpha=0.0)
    popt = fit_gaus(hist, edges, mu=np.mean(dphi), sig=np.std(dphi))
    # Redo for display
    hist, edges, _ = ax.hist(dphi, bins=600/4, range=(-300,300), histtype='step', color=color, linewidth=1.0)
    ax.text(0.05, 0.95-0.052*j, '[{0:d},{1:d}] cm: {2:.2f} +/- {3:.2f}'.format(d0_low, d0_high, popt[1], popt[2]), color=color,
            fontsize=8, transform=ax.transAxes)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_yscale('log')
    ax.set_xlim(-300, 300)
    ax.set_ylim(1.5, 8e3)

plt.show()
```
