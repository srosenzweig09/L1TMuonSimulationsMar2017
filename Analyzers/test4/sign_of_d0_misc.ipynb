```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def my_sign(x):
  return 1.0 * (x >= 0.) - 1.0 * (x < 0.)
```

### Misc (1)

```python
xv, yv, zv = 0.0014285, -0.000187, 0.5513316
eta = 0.2
phi = 45
invPt = -1./3

phi_rad = np.deg2rad(phi)
R = -(1.0/(0.003 * 3.8 * invPt))
xc, yc = xv - R * np.sin(phi_rad), yv + R * np.cos(phi_rad)

print xv, yv, xc, yc, R
```

```python
from matplotlib.patches import Circle
vertex = Circle((xv, yv), 3, color='b')
center = Circle((xc, yc), 3, color='r')
circ = Circle((xc, yc), R, color='r', fill=False)

"""
***********************************************************
*    Row   * Instance * vh_sim_r  * vh_sim_r  *  vh_sim_z *
***********************************************************
*        0 *        0 * 14.847969 * 16.108282 * 4.9522600 *
*        0 *        1 * 24.674367 * 28.409568 * 8.0585594 *
*        0 *        2 * 32.151066 * 38.915103 *  10.47612 *
*        0 *        3 * 40.988518 * 52.981149 *   12.6545 *
*        0 *        4 * 50.718096 * 71.239112 * 17.679500 *
*        0 *        5 * 59.568498 * 91.716253 * 22.736198 *
"""

circ1 = Circle((14.847969, 16.108282), 3, color='k')
circ2 = Circle((24.674367, 28.409568), 3, color='k')
circ3 = Circle((32.151066, 38.915103), 3, color='k')
circ4 = Circle((40.988518, 52.981149), 3, color='k')
circ5 = Circle((50.718096, 71.239112), 3, color='k')
circ6 = Circle((59.568498, 91.716253), 3, color='k')

fig, ax = plt.subplots(figsize=(8,8))
ax.add_artist(vertex)
ax.add_artist(center)
ax.add_artist(circ)
ax.add_artist(circ1)
ax.add_artist(circ2)
ax.add_artist(circ3)
ax.add_artist(circ4)
ax.add_artist(circ5)
ax.add_artist(circ6)
ax.axhline(y=0, alpha=0.5, color='black')
ax.axvline(x=0, alpha=0.5, color='black')
ax.set_xlim((-300, 300))
ax.set_ylim((-300, 300))

plt.show()
```

```python
for z in [4.9522600, 8.0585594, 10.47612, 12.6545, 17.679500, 22.736198]:
  cot = np.sinh(eta)
  arg_term = (z - zv)/cot
  sin_term = (2 * R) * np.sin(arg_term/(2 * R))
  cos_term = (2 * R) * (1 - np.cos(arg_term/(2 * R)))
  x = xv + np.cos(phi_rad) * sin_term - np.sin(phi_rad) * cos_term
  y = yv + np.sin(phi_rad) * sin_term + np.cos(phi_rad) * cos_term
  #print(xv + np.cos(phi_rad) * sin_term, np.sin(phi_rad) * cos_term, 
  #      yv + np.sin(phi_rad) * sin_term, np.cos(phi_rad) * cos_term)
  print(x, y, np.hypot(x, y))
```

### Misc (2)

```python
xv, yv = 50, 50
phi = 135
invPt = -0.5

phi_rad = np.deg2rad(phi)
R = -(1.0/(0.003 * 3.8 * invPt))
xc, yc = xv - R * np.sin(phi_rad), yv + R * np.cos(phi_rad)

R1 = -1 * R
xc1, yc1 = xv - R1 * np.sin(phi_rad), yv + R1 * np.cos(phi_rad)

d0 = R - my_sign(R) * np.sqrt(xc*xc + yc*yc)
xd, yd = d0 * np.sin(phi_rad), -d0 * np.cos(phi_rad)
assert(my_sign(d0) == my_sign(xv*np.sin(phi_rad) - yv*np.cos(phi_rad)))

d0_1 = R1 - my_sign(R1) * np.sqrt(xc1*xc1 + yc1*yc1)
xd1, yd1 = d0_1 * np.sin(phi_rad), -d0_1 * np.cos(phi_rad)
assert(my_sign(d0_1) == my_sign(xv*np.sin(phi_rad) - yv*np.cos(phi_rad)))

print R, xc, yc, d0, xd, yd
print R1, xc1, yc1, d0_1, xd1, yd1
```
